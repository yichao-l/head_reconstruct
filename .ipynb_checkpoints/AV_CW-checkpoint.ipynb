```python
img1 = cv2.imread('head_3_5.png',0)
img2 = cv2.imread('head_3_6.png',0)
```

```python
kp1, des1 = HEAD_RECON.estimate_frame_transform.get_descriptors("head_3_5.png")
kp2, des2 = HEAD_RECON.estimate_frame_transform.get_descriptors("head_3_6.png")
```

```python
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1,des2, k=2)
```

```python
# Apply ratio test
good = []
for m,n in matches:
    if m.distance < 0.75*n.distance:
        good.append([m])

# cv2.drawMatchesKnn expects list of lists as matches.
img3 = cv2.drawMatchesKnn(img1,kp1,img2,kp2,good,None,flags=2)

plt.imshow(img3),plt.show()
plt.imsave("des_match.png",img3)
```
